# 编码定类变量

```python
from sklearn.base import TransformerMixin
```

```python
# 自定义虚拟变量编码器
```

```python
class CustomDummifier(TransformerMixin):
    def __init__(self,cols=None):
        self.cols = cols
    
    def transform(self,X):
        return pd.get_dummies(X,columns=self.cols)
    
    def fit(self,*_):
        return self
```

```python
# 实例化自定义虚拟化器
```

```python
cd = CustomDummifier(cols = ['SFDBRDBQY'])
```

```python
df = cd.fit_transform(df)
```

```python
df = df.rename(columns={'SFDBRDBQY_否':'SFDBRDBQY_F', 'SFDBRDBQY_是':'SFDBRDBQY_T'})
df
```

# 多元线性回归

```python
pd_data = pd.read_excel('./data/mlrm_data.xlsx')
pd_data
```

```python
import seaborn as sns
```

```python
sns.pairplot(pd_data, x_vars=['X1','X2','X3','X4','X5'], y_vars='Y',kind="reg", size=5, aspect=0.7)
plt.show()
```

## 建模

```python
# 小写的 ols 函数才会自带截距项，OLS 则不会
from statsmodels.formula.api import ols
```

```python
# 固定格式：因变量 ~ 自变量(+ 号连接)
lm = ols('Y ~ X1 + X2 + X3 + X4 + X5', data=pd_data).fit()
lm.summary()
```

## 模型优化

```python
lm2 = ols('Y ~ X2 + X4 + X5', data=pd_data).fit()
lm2.summary()
```

## 基于回归模型识别异常点

```python
outliers = lm2.get_influence()
# 高杠杆值点（帽子矩阵）
leverage = outliers.hat_matrix_diag
# diffits值
dffits = outliers.dffits[0]
# 学生化残差
resid_stu = outliers.resid_studentized_external
# cook距离
cook = outliers.cooks_distance[0]
# 合并各种异常值检验的统计量值
contat1 = pd.concat([pd.Series(leverage, name='leverage'), pd.Series(dffits, name='dffits'),
                     pd.Series(resid_stu, name='resid_stu'), pd.Series(cook, name='cook')], axis=1)
```

```python
# 重设data数据的行索引
pd_data.index = range(pd_data.shape[0])
```

```python
# 将上面的统计量与train数据集合并
profit_outliers = pd.concat([pd_data, contat1], axis=1)
print(profit_outliers.head())
```

```python
# 为了简单起见，这里使用标准化残差，当标准化残差大于2时，即认为对应的数据点为异常点
outliers_ratio = sum(np.where((np.abs(profit_outliers.resid_stu) > 2),1,0)) / profit_outliers.shape[0]
print(outliers_ratio)
```

```python
# 异常比例不高，低于5%，可以考虑删除
# 挑选非异常观测点
none_outliers = profit_outliers.loc[np.abs(profit_outliers.resid_stu) <= 2, :]
```

```python
# 应用无异常值的数据集重新建模
lm3 = ols('Y ~ X2 + X4 + X5', data=none_outliers).fit()
print(lm3.params)
```

## 模型预测

```python
data_test = pd.read_excel('./data/mlrm_test_data.xlsx')
data_test
```

```python
pred = lm3.predict(data_test[['X2','X4','X5']])
```

```python
data_test['PRED'] = pred
data_test
```

## 模型评估

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,explained_variance_score
```

```python
print('测试集R~2：',round(r2_score(data_test['Y'],pred),4))
print('测试集均方误差：',round(mean_squared_error(data_test['Y'],pred),4))
print('测试集解释方差分：',round(explained_variance_score(data_test['Y'],pred),4))
print('测试集绝对误差：',round(mean_absolute_error(data_test['Y'],pred),4))
```

## 真实值与预测值对比图

```python
plt.figure(figsize = (15,5))
plt.plot(data_test['Y'],label = "真实值",color = "r",marker = "o")
plt.plot(pred,label = "预测值",color = "g",marker = "o")
plt.xlabel("测试集数据序号")
plt.ylabel("数据值")
plt.legend()
```
